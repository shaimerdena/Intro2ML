```python
import pandas as pd
```

### 1. Load the (SuperMarket Analysis.csv) dataset and display the first 5 rows.

```python
path = "SuperMarket Analysis.csv"

sales = pd.read_csv(path)

sales.head(5)
```

### 2. Display the dataset shape (rows and columns).

```python
sales.shape
```

### 3. List all column names.

```python
sales.columns
```

### 4. Identify categorical and numerical columns.

```python
categorical_cols = sales.select_dtypes(include="object").columns
numerical_cols = sales.select_dtypes(exclude="object").columns

print(categorical_cols)
print(numerical_cols)
```

### 5. Check for missing values in each column.

```python
sales.isnull().sum()
```

### 6. Display the data types of each column.

```python
sales.dtypes
```

### 7.Show summary statistics for numerical columns (use pandas method).

```python
sales.describe()
```

### 8.Filter rows where Sales > 500.

```python
sales[sales['Sales'] > 500]
```

### 9. Filter sales in City = “Yangon” and Sales > 200.

```python
sales[(sales['City'] == 'Yangon') & (sales['Sales'] > 200)]
```

### 10. Sort all orders by Sales in descending order.

```python
sales.sort_values(by = 'Sales', ascending = False)
```

### 11. Sort by Date (ascending) and then Time (ascending).

```python
sales.sort_values(by = ['Date', 'Time'], ascending = [True, False])
```

### 12.Sort by Unit price and Quantity.

```python
sales.sort_values(by = ['Unit price', 'Quantity'], ascending = [True, True])
```

### 13. Calculate the total sales in “Sales” per Branch.

```python
sales.groupby('Branch')['Sales'].sum()
```

### 14. Calculate average Sales per City.

```python
sales.groupby('City')['Sales'].mean()
```

### 15. Find the quantities sold per product line.

```python
sales.groupby('Product line')['Quantity'].sum()
```

### 16.Calculate average gross income per Gender.

```python
sales.groupby('Gender')['gross income'].mean()
```

### 17. Count number of sales per Payment method.

```python
sales.groupby('Payment')['Sales'].count()
```

### 18.Find maximum Sales per Branch.

```python
sales.groupby('Branch')['Sales'].max()
```

### 19. Find minimum Unit price per Product line.

```python
sales.groupby('Product line')['Unit price'].min()
```

### 20. Find the sum of gross income per Product line and Branch.

```python
sales.groupby(['Product line', 'Branch'])['gross income'].sum()
```

### 21. What is the total quantities sold in Product line: “Electronic accessories”?

```python
sales.groupby('Product line')['Quantity'].sum()['Electronic accessories']
```

### 22. What is the average Sales for female customers?

```python
sales.groupby('Gender')['Sales'].mean()['Female']
```

### 23. What is the most expensive Unit price among Customer type members only?

```python
sales.groupby('Customer type')['Unit price'].max()['Member']
```

### 24. How many orders with Rating >= 9 ?

```python
sales[sales['Rating'] >= 9.0]['tip'].count()
```

### 25. What is the total Sales for Payment “Credit card” in Branch Cairo?

```python
sales.groupby(['Branch', 'Payment'])['Sales'].sum()['Cairo', 'Credit card']
```
