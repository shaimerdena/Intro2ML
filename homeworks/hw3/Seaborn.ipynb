## Q1: Loading and Understanding the Dataset

```python
# import library
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Q1.1: Load dataset
df = sns.load_dataset('tips')
```

```python
# Q1.2: display first 5 rows
df.head(5)
```

```python
# Q1.3: dataset shape and columns
print(df.shape)
print(df.columns)
```

## Q2: Basic Filtering – Who Tips More?

```python
# Q2.1: Weekend
weekend = df[df['day'].isin(['Sun', 'Sat'])]

# Weekday
weekdays = df[~df['day'].isin(['Sun', 'Sat'])]
```

```python
# Q2.2: average tip

print('Average tip on weekends: ', weekend['tip'].mean())
print('Average tip on weekdays: ', weekdays['tip'].mean())
print('Average total bill on weekends: ', weekend['total_bill'].mean())
print('Average total bill on weekdays: ', weekdays['total_bill'].mean())
```

#### Q2.3: Comparison between weekday and weekend customers.
Average tips are higher on weekends than weekdays.
Average total bills are higher on weekends than weekdays. 

## Q3: Sorting and Identifying High-Value Customers

```python
# Q3.1: Sort customers by total_bill (descending).

df.sort_values('total_bill', ascending = False)
```

```python
# Q3.2: Display the top 10 highest bills.

df.sort_values('total_bill', ascending = False).head(10)
```

#### Q3.3: What do you notice about their group sizes?

Their group sizes are more than 2 and mostly equal to 4, which means they order more food.

## Q5: Visualization – What Influences Tips?

```python
#Q5.1: Tip vs Total Bill (scatter plot)

sns.scatterplot(x = "tip", y = "total_bill", data = df, alpha=1, s=50)
plt.title('Tip vs Total Bill')
plt.show()
```

```python
# Q5.2: Average tip by day (bar plot)

avg_tip_by_day = df.groupby('day',observed=False)['tip'].mean().reset_index()
sns.barplot(x='day', y='tip', data=avg_tip_by_day)
plt.title("Average Tip by Day")
plt.show()
```

```python
# Q5.3: Tip distribution (histogram)

sns.histplot(df['tip'], kde = True, bins = 30)
```

#### Q5.4: Interpret each plot in 1–2 sentences.
5.1: the scatter plot shows the dependence between tip and total_bill. According to the graphic, there is a positive relationship between total bill and tip amount, meaning that customers tend to leave higher tips when their total bill increases. 
<br>
5.2: the bar plot shows the differences between days in tipping, customers tend to tip more on weekends rather compared to weekdays.
<br>
5.3: 
The histogram shows that most tips are concentrated between 2-4$, while very large tips are relatively rare.

- when tips are highest?
- how this insight could help restaurant management.

Customers generally leave higher tips on weekends than on weekdays, with the peak tipping on Saturday and Sunday. Higher bill amounts are typically associated with larger tips. Additionally, tables with more guests tend to produce higher total bills.
