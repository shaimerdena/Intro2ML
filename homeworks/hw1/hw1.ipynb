## Task 1: Create Normal Headings

Write Markdown to create:
- A level 1 heading: "My Title"
- A level 2 heading: "My Subtitle"
- A level 3 heading: "My Section"

### My Section

## Task 2: Center a Heading

Create a centered heading: "Welcome to My Notebook".

## Task 3: Insert an Image

Insert an image using the Markdown syntax.

## Task 4: Resize an Image

Resize an image to 200px width using HTML inside Markdown.
