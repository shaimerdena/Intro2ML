## Question 3:

Let the 2D vector be:

v = np.array([4, 5])

Do the following in one plot:
1. Plot the vector as an arrow or line segment from the origin (0,0) to (4,5) .
2. Add dashed x- and y-axes similar to the one in the material provided (horizontal and
vertical dashed lines through the origin).
3. Set the plot limits so the vector is clearly visible (choose reasonable limits

````python
v = np.array([4,5])
plt.plot([0, v[0]], [0, v[1]])
plt.plot([-6, 6], [0,0], 'k--')
plt.plot([0,0], [-6, 6], 'k--')
plt.axis((-6, 6, -6, 6))
plt.grid()
````

## Question 4:

Let the 3D vector be:

v3 = np.array([6, -5, 4])

Do the following in one plot:
1. Create a 3D plot and draw the vector from the origin to ([6, -5, 4]) .
2. Label the axes as x , y , z .

````python
v3 = np.array([6, -5, 4])
plt.figure(figsize = plt.figaspect(1))
plt.axes(projection='3d')
plt.plot([0, v3[0]], [0, v3[1]], [0, v3[2]])

plt.plot([0,0], [-5,5], [0,0], 'r--')
plt.plot([0,0], [0,0], [-6,6], 'r--')
plt.plot([-5,5], [0,0], [0,0], 'r--')

plt.show()
````

## Question 5

Given:

v1 = np.array([4, -2])

v2 = np.array([3, 5])

5.1. Compute v3 = v1 + v2 and print v3

````python
v1 = np.array([4, -2])
v2 = np.array([3, 5])

v3 = v1 + v2

plt.plot([0, v1[0]], [0, v1[1]], 'k', label = 'v1')
plt.plot([0, v2[0]] + v1[0], [0, v2[1]] + v1[1], label = 'v2')
plt.plot([0, v3[0]], [0, v3[1]], 'g', label = 'v3')

plt.plot([-8,8], [0,0], 'k--')
plt.plot([0,0], [-8,8], 'k--')
plt.grid()

plt.legend()
````
