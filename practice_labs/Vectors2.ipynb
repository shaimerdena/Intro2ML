## Q1. Create a Python list representing vector v = [2,-1, 3] and multiply it by scalar 4.

```python
v = [2, -1, 3]
sc = 4

sc * v
```

Explanation: the result of multiplication is incorrect from the linear algebra perspective. because, according to the linear algebra rule: scalar number should be multiplied to each element of the vector instead of expanding the vector.

```python
#converting the vector into numpy array

v = np.array(v)
sc * v
```

## Q2. Given vector a = [5,-3] and scalar $\lambda$ =-2:

```python
a = np.array([5, -3])
lmbd = -2

a_lmbd = a*lmbd
a_lmbd
```

```python
#comparing the madnitudes of the vectors

length_a = np.linalg.norm(a)
length_a_lmbd = np.linalg.norm(a_lmbd)
print(length_a)
print(length_a_lmbd)
```

Explanation: $\lambda$ extended the magnitude of the initial vector and changed the direction to the opposite side.

## Q5. Given vector a = [1,2,3,4,5,6], compute its Euclidean norm manually and using NumPy.

```python
a = [1,2,3,4,5,6]

mult = np.multiply(a, a)
summation = np.sum(mult)
norm1 = np.sqrt(summation)
print("manually: ", norm1)

norm2 = np.linalg.norm(a)
print("numpy: ", norm2)
```

## Q7. Compute the dot product of vectors

a = [1,2,3,4,5,6] & b = [0,-4,3,-3,6,5]

```python
a = np.array([1,2,3,4,5,6])
b = np.array([0,-4,3,-3,6,5])

ab1 = 0
#manual loop
for i in range(len(a)):
    ab1 += a[i] * b[i]
print(ab1)

#numpy.dot
ab2 = np.dot(a, b)
print(ab2)
```
