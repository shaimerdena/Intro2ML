## Data Loading & Exploration

```python
#Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
#To suppress all warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#1 Load the Titanic dataset using seaborn
df = sns.load_dataset('titanic')
```

```python
#Display the first 10 rows.

df.head(10)
```

```python
#3 Count missing values in each column

df.isnull().sum()
```

```python
#4 Display column names.

df.columns
```

```python
#5 Show unique values in the embarked column.

df['embarked'].unique()
```

## Data cleaning

```python
#6 Fill missing embarked with the most frequent value.

mode_embarked = df.embarked.mode()[0]
print(mode_embarked)
df.embarked.fillna(mode_embarked)

df.isnull().sum()
```

```python
#7 Convert sex and class to categorical type.

df['sex'] = df['sex'].astype('category')
df['class'] = df['class'].astype('category')

df.info()
```

```python
# Create a new column family_size= sibsp + parch + 1

df['family_size'] = df['sibsp'] + df['parch'] + 1

df.info()
```

```python
# Cap fare at the 99th percentile to reduce outlier impact.

p_99 = df['fare'].quantile(0.99)
df['fare_capped'] = df['fare'].clip(upper = p_99)

df.describe()
```

## Visualization

```python
# 1. Pairplot of ['age', 'fare', 'sibsp'] colored by survived.

sns.pairplot(df, vars = ['age', 'fare', 'sibsp'], hue = "survived", diag_kind = "kde")

plt.show()
```

```python
# Heatmap of correlations for numeric columns.

plt.figure(figsize = (7,7))
numeric_cols = df.select_dtypes(include = np.number).columns
sns.heatmap(df[numeric_cols].corr(), annot = True, cmap = "coolwarm", center = 0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter plot of age vs fare, colored by survived.

sns.scatterplot(x = "age", y = "fare_capped", hue = "survived", style = "sex",
               alpha = 0.7, data = df)
plt.title('Age vs Fare by Survival and Sex')
plt.show()
```

```python
# Plot a histogram of age with 30 bins and KDE

sns.histplot(df['age'], kde = True, bins = 30)
plt.title('Age distribution')
plt.show()
```
