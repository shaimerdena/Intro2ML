### Kerala Flood Data

Data of monthly rainfall in Kerala from 1901 to 2018 is provided. It also contains information of floods.

### Columns in the Dataset:
- YEAR: Year between 1901 to 2018
- JAN, FEB,..., DEC:Rainfall in mm in the specified month
- ANNUAL RAINFALL: Total rainfall in mm in that year
- FLOODS: YES or NO, in that year

```python
df = pd.read_csv('kerala_floods.csv')
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

### We need to use Bays Therorm to calculate the probability of having a rainfall more than 500 mm in month of June, provided it has flooded in that year ?

1 - We may start by preparing the required columns:
- Create a new column named "June_more_than_500" and assign a value of 1 if the amount is more than 500 and 0 if it is less
than 500.
- Do the same for the 'FLOODS' column, assigning 1 for 'yes' and 0 for 'no'.

```python
df['June_more_than_500'] = (df['JUL']>500).astype('int')
df['FLOODS'] = df['FLOODS'].map({"YES":1,"NO":0})
df.head()
```

2 - Select only the columns that will help to answer the main question.

```python
df2 = df[['YEAR', 'FLOODS', 'June_more_than_500']]
df2.head()
```

3- Create a cross tabulation for 'FLOODS' and 'June_more_than_500'

```python
pd.crosstab(df2['June_more_than_500'], df2['FLOODS'], margins = True)
```

8- What is the probability of a rainfall less than 500mm in June?

```python
p_not_A = 22/118
p_not_A
```

9- What is probability of having a flood given that it rained less than 500mm?

10- Answer this question by using Bayes Theorem and the law of the Total Probaility

```python
p_A_given_B = (p_B_given_A * p_A) / (p_B_given_A * p_A + p_B_given_not_A * p_not_A)
p_A_given_B
```
