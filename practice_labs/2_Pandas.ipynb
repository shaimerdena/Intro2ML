## 14. What is the quantity of the most expensive item ordered?

```python
chipo.item_price.sort_values(ascending = False).head(1)
```

## 15. How many times was a Veggie Salad Bowl ordered?

```python
chipo[chipo.item_name == 'Veggie Salad Bowl']['quantity'].sum() 
```

## 16. How many times did someone order more than one Canned Soda?

```python
chipo[(chipo.item_name == 'Canned Soda') & (chipo.quantity > 1)]['quantity'].count()
```

## 17. How many different products are sold?

```python
unique_products = chipo['item_name'].nunique()
print(unique_products)
```

## 18. What is the total revenue?

```python
total_revenue = (chipo['item_price']* chipo['quantity']).sum()
print(total_revenue)
```

## 19. What is the average price of items?

```python
avg_price = chipo['item_price'].mean()
print(avg_price)
```

## 20. How many orders were made in total?

```python
total_number_orders = chipo['order_id'].nunique()
print(total_number_orders)
```

## 21. What is the total quantity of items ordered?

```python
total_quantity = chipo['quantity'].sum()
print(total_quantity)
```

## 22. Which item has the highest average price?
