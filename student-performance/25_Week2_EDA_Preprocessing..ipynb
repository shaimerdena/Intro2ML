### 1. Encoding categorical columns

```python
#binary encoding

df['sex'] = df['sex'].map({'F': 0, 'M': 1})               #Female - 0, Male - 1
df['school'] = df['school'].map({'GP': 0, 'MS': 1})       #Gabriel Pereira - 0, Mousinho da Silveira - 1
df['address'] = df['address'].map({'U': 0, 'R': 1})       #urban - 0, rural - 1
df['famsize'] = df['famsize'].map({'LE3': 0, 'GT3':1})    #less than 3 - 0, greater than 3 - 1
df['Pstatus'] = df['Pstatus'].map({'T': 0, 'A': 1})       #living together - 0, apart - 1
df['schoolsup'] = df['schoolsup'].map({'yes': 1, 'no': 0}) 
df['famsup'] = df['famsup'].map({'yes': 1, 'no': 0})
df['paid'] = df['paid'].map({'yes': 1, 'no': 0})
df['activities'] = df['activities'].map({'yes': 1, 'no': 0})
df['nursery'] = df['nursery'].map({'yes': 1, 'no': 0})
df['higher'] = df['higher'].map({'yes': 1, 'no': 0})
df['internet'] = df['internet'].map({'yes': 1, 'no': 0})
df['romantic'] = df['romantic'].map({'yes': 1, 'no': 0})
```

##### Notes: Binary categorical variables were encoded as 0/1, where 1 indicates the presence of a feature.

```python
# one-hot encoding

df = pd.get_dummies(df, columns=['Mjob', 'Fjob', 'reason', 'guardian'], drop_first=True).astype(int)
```

##### Notes:  Nominal categorical variables without an inherent order (Mjob, Fjob, reason, guardian) were transformed using one-hot encoding.

### Summary
<h4>
<br>Binary categorical features were mapped to 0/1. <br> Nominal categorical features without a natural order were converted into separate columns for each category using one-hot encoding to avoid introducing artificial ordinality. <br>Ordinal and numerical features were kept as integers. <br>These preprocessing steps ensure the dataset is ready for modeling.
